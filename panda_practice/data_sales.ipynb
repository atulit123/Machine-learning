```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('data_sales.xlsx')
```

```python
df2=pd.pivot_table(df,index=['year'])
```

```python
df3=pd.pivot_table(df,index=['year'],values=['HTC','sony'])
f = plt.figure()

plt.title('Title here!', color='black')
df3.plot(kind='bar' ,ax=f.gca())
lgd=plt.legend(loc='center left', bbox_to_anchor=(-0.6, 0.5))
f.savefig('samplefigure', bbox_extra_artists=(lgd,), bbox_inches='tight',dpi=200)
plt.show()
```

```python
df1_curr=pd.pivot_table(df,index=['year'],values=['TCL communication'])
print df1_curr
f1=plt.figure(2)
plt.title("Data for "+'sony')
df1_curr.plot(ax=f1.gca())
lgd1=plt.legend(loc='center left', bbox_to_anchor=(-0.6, 0.5))
f1.savefig('temp1.png', bbox_extra_artists=(lgd1,), bbox_inches='tight')

f2=plt.figure(3)
plt.title("bar graph for "+"sony")
df1_curr.plot(kind='bar',ax=f2.gca())
lgd2=plt.legend(loc='center left', bbox_to_anchor=(-0.6, 0.5))
f2.savefig('temp2.png', bbox_extra_artists=(lgd2,), bbox_inches='tight')
plt.show()
```
