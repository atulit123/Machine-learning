```python
##applying bins to age
df2['Age'].describe()
```

```python
bins=[0,15,30,45,60]
labels=['Unknown','child','young','adult','old','senior']
minval=df2['Age'].min()
maxval=df2['Age'].max()
break_points=[minval]+bins+[maxval]
df2['Age_bin']=pd.cut(df2['Age'],bins=break_points,labels=labels,include_lowest=True)
```

```python
df2['Age_bin'].describe()
```

```python
df2['Age_bin'].value_counts()
```

```python
temp=pd.pivot_table(df2,index=['Age_bin'],values=['Survived'],aggfunc=lambda x:x.mean())
```

```python
temp
```

```python
temp=pd.crosstab(df2.Age_bin,df2.Survived)
```

```python
temp1=temp.apply(lambda x:x/float(x.sum()),axis=1)
temp1
```

```python
temp1.plot(kind='bar',stacked=False)
```
