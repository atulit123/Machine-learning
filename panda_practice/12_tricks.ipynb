# Imputing Missing data

```python
from scipy.stats import mode
```

```python
mode(df['Gender'])
```

```python
mode(df['Married'])
```

```python
df['Gender'].fillna(mode(df['Gender']).mode[0],inplace=True)
df['Married'].fillna(mode(df['Married']).mode[0],inplace=True)
df['Self_Employed'].fillna(mode(df['Self_Employed']).mode[0],inplace=True)
```

```python
a=df.apply(lambda x:sum(x.isnull()),axis=0)
a
```

# Multi Indexing

```python
impute_grps=pd.pivot_table(df,index=['Gender','Married','Self_Employed'],values=['LoanAmount'],aggfunc=lambda x:x.mean())
```

```python
impute_grps
```

```python
impute_grps.loc[('Male','No','No')]
```

```python
df[df['LoanAmount'].isnull()]
    
```

```python
for i,row in df[df['LoanAmount'].isnull()].iterrows():
    ind=tuple([row['Gender'],row['Married'],row['Self_Employed']])
    df.loc[i,'LoanAmount']=impute_grps.loc[ind].values[0]
```

```python
df.apply(lambda x:sum(x.isnull()),axis=0)
```

# Binning

```python
df.LoanAmount.describe()
```

```python
bins=[90,140,190]
labels=['low','medium','high','very high']
minval=df['LoanAmount'].min()
maxval=df['LoanAmount'].max()
break_points=[minval]+bins+[maxval]
df['LoanAmount_bin']=pd.cut(df['LoanAmount'],bins=break_points,labels=labels,include_lowest=True)
```

```python
df['LoanAmount_bin'].value_counts()
```
